### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


```python
# 防止GitHub無法顯示
with open("./data/64_72hr_CH.xml", "r",encoding="UTF-8") as f:
    fr = f.read()
# print(fr)
```

```python
import xmltodict

d = dict(xmltodict.parse(fr))
# d
```

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

import xmltodict

data = []
d = dict(xmltodict.parse(fr))
locs = d['cwbopendata']['dataset']['locations']['location']
for loc in locs:
    data.append(loc['weatherElement'])
print('高雄市有%d個地區有溫度資料'%len(data))
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

data = []
locs = d['cwbopendata']['dataset']['locations']['location']
for loc in locs:
    for w_element in loc['weatherElement']:
        if(w_element['description']=='溫度'):
            data.append([loc['locationName'],w_element['time'][0]['dataTime'],w_element['time'][1]['elementValue']['value']])
data
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

data = []
locs = d['cwbopendata']['dataset']['locations']['location']
for time in locs[0]['weatherElement'][0]['time']:
    data.append([time['dataTime'],time['elementValue']['value']])
print(locs[0]['locationName'])
for i in range(len(data)):
    print(data[i-1])
```
