## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests
import json

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
j = json.loads(r.text)
j
```

```python
count = 0
for data in j:
    count = count +1
print('有%d筆資料'%count)
```

```python
print('每筆資料有:')
for data_col in j[0]:
    print(data_col)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for data in j:
    print('標題:',data['title'],'貼文時間:',data['updatedAt'],'留言人數:',data['commentCount'],'按讚人數:',data['likeCount'])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
j = json.loads(r.text)
count = 0
csum = 0
lsum = 0
for data in j:
    csum = csum + int(data['commentCount'])
    lsum = lsum + int(data['likeCount'])
    count = count + 1
print('熱門文章平均留言人數:%d'%(int(csum/count)))
print('熱門文章平均案讚人數:%d'%(int(lsum/count)))
```

```python
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
j = json.loads(r.text)
count = 0
csum = 0
lsum = 0
for data in j:
    csum = csum + int(data['commentCount'])
    lsum = lsum + int(data['likeCount'])
    count = count + 1
print('非熱門文章平均留言人數:%d'%(int(csum/count)))
print('非熱門文章平均案讚人數:%d'%(int(lsum/count)))
```
