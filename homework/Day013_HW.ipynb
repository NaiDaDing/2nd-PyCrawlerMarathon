## 作業目標

根據範例 ，完成以下問題：

* ① 印出最新文章的「作者」「標題」「時間」
* ② 印出第一頁所有文章的「作者」「標題」「時間」


```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ptt.cc/bbs/NBA/index.html'
r = requests.get(url)

soup = BeautifulSoup(r.text, 'html5lib')
soup
```

### ① 印出最新文章的「作者」「標題」「時間」

```python
lastest_dict = {
    'author': [], 
    'title': [],
    'time': []
}

lastest_dict['author'].append(soup.find(class_='meta').find(class_="author").text)
lastest_dict['title'].append(soup.find(class_='title').find('a').text)
lastest_dict['time'].append(soup.find(class_='meta').find(class_="date").text)

lastest_dict
```

### ② 印出第一頁所有文章的「作者」「標題」「時間」

```python
for data in soup.find_all(class_='r-ent'):
    try:
        author = data.find(class_='meta').find(class_="author").text
        title = data.find(class_='title').find('a').text
        time = data.find(class_='meta').find(class_="date").text
    except Exception as e:
        continue
    print('Author:', author, '\nTitle:', title, '\nTime:', time)
```
